```python
"""
1.Merge Dataframes
"""
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
print(df1)
print('===========================================================')

df2 = pd.DataFrame({ "city": ["chicago","new york","orlando"], "humidity": [65,68,75], }) 
print(df2)
print('===========================================================')
df3 = pd.merge(df1, df2, on="city")
print(df3)
print('===========================================================')
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
print(df1)
print('===========================================================')
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
print(df2)
print('===========================================================')
df3=pd.merge(df1,df2,on="city",how="inner")
print(df3)
print('===========================================================')
df3=pd.merge(df1,df2,on="city",how="outer")
print(df3)
print('===========================================================')
df3=pd.merge(df1,df2,on="city",how="left")
print(df3)
print('===========================================================')

df3=pd.merge(df1,df2,on="city",how="right") 
print(df3)
print('===========================================================')
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
print(df3)
print('===========================================================')
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
print(df1)
print('===========================================================')

df2 = pd.DataFrame({ "city": ["chicago","new york","san diego"], "temperature": [21,14,35], "humidity": [65,68,71] }) 
print(df2)
print('===========================================================')
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
print(df3)

print('===========================================================')
df1 = pd.DataFrame({ "city": ["new york","chicago","orlando"], "temperature": [21,14,35], }) 
df1.set_index('city',inplace=True) 
print(df1)
print('===========================================================')
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})

df2.set_index('city',inplace=True)
print(df2)
print('===========================================================')

df1.join(df2,lsuffix='_l', rsuffix='_r')
```

```python
"""
2.Pivot-redefine col and row
"""
import pandas as pd
import numpy as np
%cd drive/MyDrive/Colab Notebooks
df = pd.read_csv("weather.csv")
print(df)
print('===========================================================')

print(df.pivot(index='city',columns='date'))
print('===========================================================')
print(df.pivot(index='city',columns='date',values="humidity"))
print('===========================================================')
print(df.pivot(index='date',columns='city'))
print('===========================================================')
print(df.pivot(index='humidity',columns='city'))

print('===========================================================')
df = pd.read_csv("weather2.csv") 
print(df)
print('===========================================================')
print(df.pivot_table(index="city",columns="date"))
print('===========================================================')
print(df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum))
# margins = True: get ave. value shown in all

df = pd.read_csv("weather3.csv")
print(df)
print('===========================================================')
df['date'] = pd.to_datetime(df['date'])
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```

```python
"""
4.Stack Unstack
"""
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1])
print(df)
print('===========================================================')
print(df.stack())
print('===========================================================')
print(df.stack(level=0))
print('===========================================================')
df_stacked=df.stack()
print(df_stacked)
print('===========================================================')
print(df_stacked.unstack())
print('===========================================================')
df2 = pd.read_excel("stocks_3_levels.xlsx",header=[0,1,2])
print(df2)
print('===========================================================')
print(df2.stack())
print('===========================================================')
print(df2.stack(level=0))
print('===========================================================')
print(df2.stack(level=1))
```

```python
"""
5.Crosstab
"""
import pandas as pd 
df = pd.read_excel("survey1.xlsx") 
print(df)
print('===========================================================')
print(pd.crosstab(df.Nationality,df.Handedness))
print('===========================================================')
print(pd.crosstab(df.Sex,df.Handedness))
print('===========================================================')
print(pd.crosstab(df.Sex,df.Handedness, margins=True))
print('===========================================================')
print(pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True))
# marigins: get summation
print('===========================================================')
print(pd.crosstab([df.Nationality, df.Sex], [df.Handedness], margins=True))
print('===========================================================')
print(pd.crosstab(df.Sex, df.Handedness, normalize='index'))
print('===========================================================')

import numpy as np 
print(pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average))
```
